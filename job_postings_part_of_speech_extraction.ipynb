```
def find_noun(x):
  """Get Just the Adjectives"""
  noun_list = []
  doc = nlp(str(x))
  for token in doc:
    if token.pos_ == 'NOUN':
      noun_list.append(token.orth_)
  return noun_list

df['nouns'] = df['job_description'].apply(lambda x: find_noun(x))
df.tail(1)
```

```
def find_adj(x):
  """Get Just the Adjectives"""
  adj_list = []
  doc = nlp(str(x))
  for token in doc:
    if token.pos_ == 'ADJ':
      adj_list.append(token.orth_)
  return adj_list

df['adj'] = df['job_description'].apply(lambda x: find_adj(x))
df.tail(1)
```

```
def find_subjects(x):
  """Get Just the Adjectives"""
  new_list = []
  sub_list = x.split(' ')
  for x in sub_list:
      if len(x) > 2:
          if x[0].isupper() == True:
              new_list.append(x)
  return new_list

df['sub'] = df['job_description'].apply(lambda x: find_subjects(x))
df['sub'].tail(3)

# if it is not a place, organization
# if it is not a verb 
```

```
# all_similar_job_links = list(df['all_similar_job_links'][0])
def get_similar_links_list(all_similar_job_links):
    recommended_links = []
    if type(all_similar_job_links) != float: 
        for similar_job in all_similar_job_links:
            start = similar_job.find('href=')
            end = similar_job.find('"')
            if start != -1 and end != -1:
                result = similar_job[start:end + 1]
                similar_job = similar_job.replace(result, '')
                end_quote = similar_job.find('"')
                rec_url = similar_job[0:end_quote]
                recommended_links.append(rec_url)
    return recommended_links

def augment(x):
    try:
        if type(x) == float:
            x = str(x)
        elif type(x) == str:
            x = x.split(' ')
        return x
    except:
        return []

df['all_similar_job_links_str'] = df['all_similar_job_links'].apply(lambda x: augment(x))
df['all_similar_job_links_str'] = df['all_similar_job_links_str'].apply(lambda x: get_similar_links_list(x))
df.tail(3)
```

```
import re

def just_applicants(x):
    app_appearance = x.rfind(' applicants')
    app = x[app_appearance-5:app_appearance]
    some_num_str = re.findall(r'\d+', app)
    days_ago = ''.join(some_num_str)
    return days_ago

df['num_applicants'] = df['posted_days_ago'].apply(lambda x: just_applicants(x))
df.tail(3)
```

```
import re

def find_days_ago(x):
    days_str = x.rfind('days ago')
    hours_str = x.rfind('hours ago')
    days = x[days_str-5:days_str]
    hours = x[hours_str-5:hours_str]
    final_day_str = re.findall(r'\d+', days)
    final_hour_str = re.findall(r'\d+', hours)
    days_ago = ''.join(final_day_str)
    final_hour_str = ''.join(final_hour_str)
    if len(final_hour_str) > 0:
        final_hour_str = str(round(int(final_hour_str)/24,2))
    result = days_ago + final_hour_str
    return result

df['posted_days_ago_cleaned'] = df['posted_days_ago'].apply(lambda x: find_days_ago(x))
df.tail(3)
```
